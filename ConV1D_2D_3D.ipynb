```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(figsize=(18, 6))

# Conv1D
ax1 = fig.add_subplot(1, 3, 1)
x = np.linspace(0, 9, 10)
y = np.sin(x)
ax1.plot(x, y, 'o-', color='blue', linewidth=3, markersize=8)
ax1.set_title("Conv1D - Data Satu Dimensi", fontsize=14)
ax1.set_xlabel("Posisi")
ax1.set_ylabel("Nilai")
ax1.grid(True)

# Conv2D
ax2 = fig.add_subplot(1, 3, 2)
data2d = np.random.rand(10, 10)
heatmap = ax2.imshow(data2d, cmap='Blues', interpolation='nearest')
ax2.set_title("Conv2D - Gambar Dua Dimensi", fontsize=14)
ax2.set_xlabel("Lebar")
ax2.set_ylabel("Tinggi")
fig.colorbar(heatmap, ax=ax2)

# Conv3D
ax3 = fig.add_subplot(1, 3, 3, projection='3d')
x, y, z = np.indices((5, 5, 5))
voxels = (x + y + z) % 2 == 0
colors = np.empty(voxels.shape, dtype=object)
colors[voxels] = 'skyblue'
ax3.voxels(voxels, facecolors=colors, edgecolor='k')
ax3.set_title("Conv3D - Data Volume / Video", fontsize=14)
ax3.set_xlabel("Lebar")
ax3.set_ylabel("Tinggi")
ax3.set_zlabel("Kedalaman")

plt.tight_layout()
plt.show()
```
